# Generator instead of Regex

code source from [stackoverflow](https://stackoverflow.com/questions/4664850/how-to-find-all-occurrences-of-a-substring).

python documentation on string.find() [here](https://docs.python.org/3/library/stdtypes.html).

```python
def findall(p, s):
    '''Yields all the positions of
    the pattern p in the string s.'''
    i = s.find(p)
    while i != -1:
        yield i
        i = s.find(p, i+1)
```

```python
sample_result = findall(t, s)
```

```python
sample_result2 = [str(i+1) for i in sample_result]
sample_result2
```

# On actual problems

```python
lines2 = lines.copy()
```

```python
lines2.pop(0) #get rid of first line
```

```python
output = []
```

```python
# on actual problem
# remember to remove trailing space

for i in range(0,len(lines2),2):
    s = lines2[i].rstrip()
    t = lines2[i+1].rstrip()
    
    result_gen = findall(t, s)
    result_lst = [str(i+1) for i in result_gen]
    result_str = ' '.join(result_lst) + '\n'
    
    output.append(result_str)
```
