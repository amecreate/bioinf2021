```python
lines=[]
with open('test1.txt') as f:
    for line in f:
        if not line.isspace():
            lines.append(line)
```

```python
import numpy as np
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
lines[1]
```

# From string to DataFrame

```python
l = lines[1:].copy()
```

```python
s = pd.Series(l)
```

```python
df = s.str.split(pat='', expand=True)
df.head()
```

```python
df = df.replace({'?': np.nan})
df.head()
```

```python
df = df.drop(columns=[0,271,272])
df.head()
```

```python
df = df.astype(int, errors='ignore', copy=True)
df.head()
```

# Separate fit data from test data

```python
df_fit = df[:200].copy(deep=True).astype(int)
df_fit.tail()
```

```python
(df_fit == 2).any().sum()
```

```python
df_test = df[200:].copy(deep=True)
df_test.tail()
```

```python
df_test.head()
```

# Expand test data 

```python
len(df_test)
```

```python
na_row = [np.nan] * 270
```

```python
df_test.index
```

```python
for i in df_test.index:
    df_test.loc[i+0.5] = na_row
```

```python
df_test = df_test.sort_index().reset_index(drop=True)
```

```python
df_test.head()
```

# Fill na value

```python
df_test2 = df_test.fillna(method='ffill').copy(deep=True)
df_test2.head()
```

```python
type(df_test2.iloc[0,0])
```

```python
# change data format from string to integer, ignore nan value
df_test2 = df_test2.astype(int, errors='ignore')
```

```python
# change 1 on odd rows to 0
df_test2[1::2] = df_test2[1::2].replace(1,0)
```

```python
# change all 2 to 1
df_test2 = df_test2.replace(2,1)
```

```python
df_test2
```

# Imputation

```python
imp = IterativeImputer(max_iter=10, random_state=0)
```

```python
%%time

imp.fit(df_fit)
```

```python
df_imp = np.round(imp.transform(df_test2))
```

```python
df_imp
```

# add two arrays together

```python
output = []
for i in range(0,len(df_imp),2):
    output.append([str(int(x+y)) for x,y in zip(df_imp[i],df_imp[i+1])])
```

```python
result = [''.join(x) + '\n' for x in output]
```

```python
result[1]
```

```python
with open('result1.txt','w') as f:
    f.writelines(result)
```
